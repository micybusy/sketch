```python
import pandas as ps
from urllib.request import urlopen
import requests
```

```python
url = "https://raw.githubusercontent.com/micybusy/sketch/main/charts.csv"
with urlopen(url) as f:
    content = ps.read_csv(url, encoding = "utf-8")

content
```

```python
entries_by_year = {}
dummy_list = []
a = 2022
for i in range(0,len(content), 5200):
    dummy_list.append(content[i:i+5200])
    entries_by_year.update({a:dummy_list})
    a = a-1
    dummy_list = []
```

```python
def getYear(a):
    return entries_by_year[a][0]
```

```python
def groupby_artist(a):
    perm = getYear(a)
    out =ps.DataFrame(perm.groupby(["artist", "song"]).size().reset_index().rename(columns={0:'count'}))
    out = out.sort_values("count", ascending = False)
    return out
```

```python
groupby_artist(2019)
```

```python
url = "https://raw.githubusercontent.com/micybusy/sketch/main/the_grammy_awards.csv"
with urlopen(url) as f:
    content_grammy = ps.read_csv(url, encoding = "utf-8")

content_grammy = content_grammy.drop(columns = ["title", "published_at", "updated_at", "workers", "img", "winner"])
content_grammy.rename(columns = {'nominee':'song'}, inplace = True)
content_grammy
```

```python
def get_nominees(y, c):
    y = str(y)
    c = str(c)
    dummy = content_grammy[content_grammy["year"].astype("str").str.contains(y)]
    return dummy[dummy["category"].astype("str").str.contains(c)]
```

```python
def matcher(year):
    counts = groupby_artist(year)
    nominees = get_nominees(year, "Record Of The Year")
    out = nominees.merge(counts, on = ["song"] )
    return out
```

```python
matcher(2019)
```
